```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Download dataset
!wget -O accidents.csv https://raw.githubusercontent.com/plotly/datasets/master/2011_february_us_airport_traffic.csv

# Load dataset
df = pd.read_csv("accidents.csv")

# Show first 5 rows
print(df.head())

# Dataset information
print(df.info())

# -----------------------------
# Bar Chart
# -----------------------------
plt.figure(figsize=(10,5))

sns.countplot(x='cnt', data=df)

plt.title("Traffic Count Distribution")

plt.xticks(rotation=90)

plt.show()

# -----------------------------
# Histogram
# -----------------------------
plt.figure(figsize=(8,5))

sns.histplot(df['cnt'], bins=20)

plt.title("Histogram of Traffic Counts")

plt.xlabel("Traffic Count")
plt.ylabel("Frequency")

plt.show()
```
